```python
import matplotlib.pyplot as plt
import numpy as np

# CSVファイルからデータを読み込み
data = np.loadtxt('simulation_data.csv', delimiter=',')

# データをプロット
for i in range(data.shape[1]):
    plt.plot(data[:, i])

plt.xlabel('Time')
plt.ylabel('Probability')
plt.title('Quantum Annealing Dynamics')
plt.show()
```
